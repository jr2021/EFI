```
import pandas as pd
import numpy as np
import glob as gl
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import plotly.graph_objects as go
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
import IPython.display as ipy
import plotly
```

```
data = pd.read_csv('nfoas_raw.csv')
data
```

```
fig = go.Figure()
fig.add_trace(go.Histogram(x=data['OA']))

fig.update_layout(xaxis_title='Class', yaxis_title='Frequency', title='General Distribution')
```

```
train_test, validate = train_test_split(data, test_size=0.33)
```

```
train, test = train_test_split(train_test, test_size=0.5)
```

```
fig = go.Figure()
fig.add_trace(go.Histogram(x=train['OA'], name='Train'))
fig.add_trace(go.Histogram(x=test['OA'], name='Test'))
fig.add_trace(go.Histogram(x=validate['OA'], name='Validate'))

fig.update_layout(xaxis_title='Class', yaxis_title='Frequency', title='Train/Test/Validation Distribution')
```

```
with open('nfoas_train.csv', 'w+') as f:
    f.write(train.to_csv())
```

```
with open('nfoas_test.csv', 'w+') as f:
    f.write(test.to_csv())
```

```
with open('nfoas_validate.csv', 'w+') as f:
    f.write(validate.to_csv())
```
