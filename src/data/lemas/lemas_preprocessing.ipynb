```
import pandas as pd
import numpy as np
import glob as gl
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import plotly.graph_objects as go
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
import IPython.display as ipy
import plotly
```

```
names = []

with open('lemas/communities.names') as f:
    for line in f.readlines():
        if line[:10] == '@attribute':
            try:
                names.append(line[11:line.index('numeric') - 1])
            except:
                names.append(line[11:line.index('string') - 1])
```

```
data = pd.read_csv('lemas/communities.data', names=names)
data
```

```
columns = list(data.columns)
columns[5], columns[-1] = columns[-1], columns[5]
data = data[columns]
data
```

```
for column in data.columns[5:]:
    data[column] = data[column].replace('?', np.nan).astype(float)
    data[column] = data[column].astype(float).fillna(data[column].mean())

data
```

```
with open('lemas_raw.csv', 'w+') as f:
    f.write(data.to_csv())
```

```
_, data = train_test_split(data, test_size=0.25)
```

```
fig = go.Figure()
fig.add_trace(go.Histogram(x=data['ViolentCrimesPerPop'], name='ViolentCrimesPerPop'))

fig.update_layout(barmode='overlay', xaxis_title='Value', yaxis_title='Frequency', title='General Distribution')
fig.update_traces(opacity=0.5)
```

```
train_test, validate = train_test_split(data.drop(columns=['state', 'county', 'community', 'communityname', 'fold']), test_size=0.33)
```

```
train, test = train_test_split(train_test, test_size=0.5)
```

```
validate
```

```
fig = go.Figure()
fig.add_trace(go.Histogram(x=train['ViolentCrimesPerPop'], name='Training Set'))
fig.add_trace(go.Histogram(x=test['ViolentCrimesPerPop'], name='Testing Set'))
fig.add_trace(go.Histogram(x=validate['ViolentCrimesPerPop'], name='Validation Set'))

fig.update_layout(barmode='overlay', xaxis_title='Value', yaxis_title='Frequency', title='General Distribution')
fig.update_traces(opacity=0.5)
```

```
with open('lemas_train.csv', 'w+') as f:
    f.write(train.to_csv())
```

```
with open('lemas_test.csv', 'w+') as f:
    f.write(test.to_csv())
```

```
with open('lemas_validate.csv', 'w+') as f:
    f.write(validate.to_csv())
```
